Data Cleansing:

```python
#remove metascore

df1 = movies.drop('metascore', axis=1)
```

```python
#remove votes 

df2 = df1.drop('votes', axis=1)
```

```python
#enter N/A for missing values 

df3 = df2.fillna("N/A")
```

```python
#replace rating with score 

df4 = df3.rename(columns = {'rating' : 'score'})
```

```python
#replace certificate with rating

df5 = df4.rename(columns = {'certificate' : 'rating'})
df5
```

```python
#replace stars with actors

df6 = df5.rename(columns = {'stars' : 'actors'})
```

```python
#keep only directors name in director field (remove all others)
df6['director'] = df6['director'].str.rsplit(', ').str.get(0)
```

```python
df6
```

```python
#remove all movies that released before 1972 and after 2022

df6['year'] = df6['year'].astype('int')
```

```python
df6 = df6[(df6['year'] >= 1972) & (df6['year'] <= 2022)]
```

After analyzing the top 5000 movies released between the years of 1972-2022, we were able to determine that the most popular genre is Drama, the most popular rating is R, the most popular director is Woody Allen, and the most popular actors are Denzel Washington, Tom Hanks, and Tom Cruise. Now that we have the above information, will this lead to the highest rated movie of all time? Let's see!

```python
#find the mean score of all movies starring Denzel Washington

denzel_washington_movies = df8[df8['actors1'].str.contains('Denzel Washington')]

mean_score_DW = denzel_washington_movies['score'].mean()
mean_score_DW
```

```python
#find the mean score of all movies starring Tom Hanks

tom_hanks_movies = df8[df8['actors1'].str.contains('Tom Hanks')]

mean_score_TH = tom_hanks_movies['score'].mean()
mean_score_TH
```

```python
#find the mean score of all movies starring Tom Cruise

tom_cruise_movies = df8[df8['actors1'].str.contains('Tom Cruise')]

mean_score_TC = tom_cruise_movies['score'].mean()
mean_score_TC
```

The mean score of all Tom Cruise movies is 7.24

```python
#find the mean score of all three actors

scores = [mean_score_TC, mean_score_TH, mean_score_DW]
mean_score_top3 = np.mean(scores)
mean_score_top3
```

The combined mean score of the top 3 actors is 7.25

```python
#find the mean score of all movies directed by Woody Allen 

woody_allen_movies = df8[df8['director'].str.contains('Woody Allen')]

mean_score_WA = woody_allen_movies['score'].mean()
mean_score_WA
```

```python
#find the mean score of all movies with an R rating 

R_movies = df8[df8['rating'].str.contains('R')]

mean_score_R = R_movies['score'].mean()
mean_score_R
```

```python
#find the mean score of all movies in the Drama genre 

drama_boolean = df8['Drama'].astype(bool)

drama_movies = df8[drama_boolean]

mean_score_Drama = drama_movies['score'].mean()
mean_score_Drama
```

The mean score of all movies in the Drama genre is 7.27

Now that we have the mean score for each, let's plug them in to our equation!

```python
# define weights
weight_director = 0.3
weight_actors = 0.3
weight_rating = 0.2
weight_genre = 0.2

#mean scores 
mean_score_top3 = 7.25
mean_score_WA = 7.19
mean_score_Drama = 7.27
highest_possible_rating = 10

possible_highest_rating = (weight_director * mean_score_WA) + 
    (weight_actors * mean_score_top3) + (weight_rating * highest_possible_rating) + 
    (weight_genre * mean_score_Drama)

possible_highest_rating
```

The highest possible rating is 7.786
